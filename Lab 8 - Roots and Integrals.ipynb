##### Exercise 8.2 - Bisection Python function

```python
def bisection(f,a,b):
    c = (a + b) / 2
    if f(a) * f(c) <= 0:
        a = a
        b = c
    else:
        a = c
        b = b
    return a, b    
```

```python
print(bisection(f,0,4))
print(bisection(f,3,4))
```

##### Exercise 8.3 - Newton's Method

```python
def newton(f, fp, x):
    return x - (f(x) / fp(x))
```

```python
for x in [-100 ,5,1000]:
    print(newton(f,fp,x))
```

##### Exercise 8.4 - Secant Method

```python
def secant(f,a,b):
    c = b - f(b) * ((b - a)/(f(b) - f(a)))
    return b, c
```

```python
print(secant(f,0,4))
print(secant(f,5,3))
```

```python
def g(x):
    return (x - 1) * (x - 4)
```

```python
x = np.linspace(0,5,100)

plt.plot(x,g(x),"-y", label="g(x)")
plt.plot(x,gp(x),"-c", label="g'(x)")
plt.legend(frameon = False)
plt.title("g and its 1st derivative")
plt.xlabel("x")
plt.ylabel("g")
plt.show()
```

##### Exercise 8.9 - Reproduce Fig. 8.9

```python
def u(v):
    return v * np.tan(v)

def ue(v0, v):
    return np.sqrt((v0 ** 2)-(v ** 2))


p = np.pi

for i in range(5):
    if i == 0:
        x0 = 0
        x1 = (1 * p) / 2
    else:
        x0 = ((2 * i - 1) * p) / 2
        x1 = ((2 + (2 * i - 1)) * p) / 2
    v = np.linspace(x0, x1, 200)
    v = v[1 : -1]
    plt.plot(v,u(v),"r--")
    plt.plot(v,ue(10, v), "b-")

plt.plot(v,u(v),"r--", label = "$u=v\ tan\ v$")
plt.plot(v,ue(10,v), "b-", label = "$u=\sqrt{v_{0}^{2}-v^2}$")
plt.legend(frameon = True)
plt.legend(loc=1)
plt.xlim(0,15)
plt.ylim(-4,12)
plt.xlabel("v")
plt.ylabel("u")
plt.show()
    

```
