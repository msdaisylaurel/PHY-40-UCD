###### Exercise 5.1 - Powers of a

```python
def powers(a,n):
    m = np.array([])               # create base array
    for i in range(n+1):
        m = np.append(m, a)        # populate the array with base number, n times
    s = np.arange(0, n + 1, 1)     # create power array
    print(m ** s)                  # operate on base array using power array
    
```

```python
powers(2,4)
```

```python
def recur(x, n):
    a = np.array([])
    for i in range(n):
        a = np.append(a, x)
        x = 2 * x + 1
    print(a)
        
```

##### Exercise 5.7 - plot of the time evolution of the logistics map for r = 2.8

```python
def logmap(p,r):
    a = []
    for i in range(40):
        p = r * p * (1 - p)
        a.insert(0, p)
    return a;

UPPER = 0
LOWER = 40
tau = 2*np.pi
x = np.linspace(LOWER ,UPPER ,40)
s = logmap(0.1, 2.8)
plt.plot(x,s,"yo", label="")
plt.xlabel("iteration")
plt.ylabel("p")
plt.title("Logistics Map for r = 2.8")
plt.legend(frameon=False)
plt.show()
    
```

##### Exercise 5.10 - chaotic bifurcation

```python
def logmap(p,r):
    a = np.array([])
    r = np.flip(r)
    for i in range(100):
        p = p * (1 - p)
        p = r[i] * p
        a = np.append(a, p)
    return a;





Nr = 100
r = np.linspace(2.6,4,Nr)
p = logmap(0.1,r)
#print(p)
p = logmap(p,r)
print(p)




UPPER = 0
LOWER = 4.0
x = np.linspace(LOWER ,UPPER ,10000)
s = p
plt.plot(x,s,"k,", label="")
plt.xlabel("r")
plt.ylabel("p")
plt.title("Logistics Map for Chaotic Bifurcation")
plt.legend(frameon=False)
plt.show()


```
